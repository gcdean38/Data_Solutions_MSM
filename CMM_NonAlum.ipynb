```python
mm21 = pd.read_excel('/Users/gcdyw9/Downloads/Donations to College media madness 2021 - copy.xlsx')
mm21['DONATION_YEAR'] = 2021
mm21.head()
```

```python
mm22 = pd.read_excel('/Users/gcdyw9/Downloads/College Media Madness 2022 Donations.xlsx')
mm22['DONATION_YEAR'] = 2022

mm22.head()
```

```python
tables = [mm21,mm22]
```

```python
mm_21_22 = pd.concat(tables)
mm_21_22.shape
```

```python
def gift_21_amt(year,amt):
    if year==2022:
        return 0
    else:
        return amt
def gift_22_amt(year,amt):
    if year==2021:
        return 0
    else:
        return amt
```

```python
cmm_total_giving = mm_21_22.groupby('FULL_NAME').agg(
    FIRST_NAME = pd.NamedAgg('FIRST_NAME','max'),
    LAST_NAME = pd.NamedAgg('LAST_NAME','max'),
    PREF_EMAIL_ADDRESS = pd.NamedAgg('PREF_EMAIL_ADDRESS','max'),
    FIRST_GIFT_YEAR = pd.NamedAgg('DONATION_YEAR','min'),
    LAST_GIFT_YEAR = pd.NamedAgg('DONATION_YEAR','max'),
    TOTAL_CMM_GIFT = pd.NamedAgg('TOTAL_GIVING','sum'),
    AVERAGE_CMM_GIFT = pd.NamedAgg('TOTAL_GIVING','mean'),
    MAX_CMM_GIFT = pd.NamedAgg('TOTAL_GIVING','max'),
    NUM_GIFT_YEARS = pd.NamedAgg('DONATION_YEAR','nunique'),
    FIRST_GIFT = pd.NamedAgg('TOTAL_GIVING','first'),
    LAST_GIFT = pd.NamedAgg('TOTAL_GIVING','last')


    
)
cmm_total_giving=cmm_total_giving.reset_index()
cmm_total_giving['GIFT_2021_AMT']=cmm_total_giving.apply(lambda x: gift_21_amt(x.FIRST_GIFT_YEAR, x.FIRST_GIFT), axis=1)
cmm_total_giving['GIFT_2022_AMT']= cmm_total_giving['TOTAL_CMM_GIFT']-cmm_total_giving['GIFT_2021_AMT']
cmm_total_giving.head()
```

```python
def wave_direction(year1,year2):
    if year1 == year2:
        return 'flat'
    elif year1 > year2:
        return 'down'
    elif year1 < year2:
        return 'up'
```

```python
cmm_total_giving['WAVE_DIRECTION']=cmm_total_giving.apply(lambda x: wave_direction(x.GIFT_2021_AMT, x.GIFT_2022_AMT), axis=1)
```
